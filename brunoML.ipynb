```python
from scipy.io import arff
from sklearn import tree
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
print('bruhh')
```

```python
#input df, remove cats, replace nan
data = arff.loadarff('IanTurnipDataset.arff')
df = pd.DataFrame(data[0])
display(df)
y = pd.DataFrame(df['categorized result'])
# y.rename("binary result")
df = df.drop(['binary result','specific result','categorized result','time','setpoint','gain','reset rate','deadband','cycle time','rate'], axis=1)
df = df.replace(np.nan,-1)
y=y.astype('int')                                                                    
```

```python
with pd.option_context('display.max_rows',None):
    display(df.loc[0:50])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df,y,test_size=.2)

from sklearn.externals import joblib
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
predictions = clf.predict(X_test)
score = accuracy_score(y_test, predictions) # 80% train 20% test
print(clf.feature_importances_)

model = joblib.dump(clf,'bruno1.joblib')



```
